```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('amazon.xlsx')
df
```

```python
month_mapping = {
    'Janeiro': 'January',
    'Fevereiro': 'February',
    'Março': 'March',
    'Abril': 'April',
    'Maio': 'May',
    'Junho': 'June',
    'Julho': 'July',
    'Agosto': 'August',
    'Setembro': 'September',
    'Outubro': 'October',
    'Novembro': 'November',
    'Dezembro': 'December'
}

# Map the Portuguese month names to English month names
df['month'] = df['month'].map(month_mapping)

# Define the desired order of the months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Set the 'Month' column as an ordered categorical column
df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True)

# Sort the DataFrame by the 'Month' column
df = df.sort_values('month') 
df
```

```python
pivot=df.pivot_table(index='month', columns='state', values='number')
pivot
```

```python
pivot.plot(kind='bar').legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Month')
plt.ylabel('Number')
plt.title('Amazon fire Visualization')
plt.show()
'''
Finding_1: Summer time (June to July) is the time that is more likely to have fire, 
while spring time (February to May) is the time that is least like to have fire.
'''
```

```python
pivot_area=df.pivot_table(index='state', values='number')
pivot_area.sort_values('number')
'''
Finding_2: Sao Paulo is the area that has the most number of fire, while Sergipe has the least number.
'''
```

```python
pivot_area=df.pivot_table(index='year', values='number')
pivot_area.sort_values('year')
pivot_area.plot(kind='line').legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Amazon fire Visualization')
plt.show()
'''
Finding_3: The number of fire is growing in general but the speed of growing is decreasing.
'''
```
