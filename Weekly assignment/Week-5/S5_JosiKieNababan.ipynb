```python
#import panda
import pandas as pd
#take dataset from excel file
dataset = pd.read_excel("amazon.xlsx")
dataset
```

```python
# create dict contains head 'total of fires forests all the time' and its value
record = {"Total of Fires Forests All The Time" : [int(dataset['number'].sum())]}
# visualize it as column
totalOfFireForestsAllTheTime = pd.DataFrame(record).style.hide()
totalOfFireForestsAllTheTime
```

```python
# create dict contains head 'for average number of fires forests per year' and its value
record = {"Average number of fires forests per year" : [int(dataset.groupby('year')['number'].sum().mean())]}
# visualize it as column
averagePerYear = pd.DataFrame(record).style.hide()
averagePerYear
```

```python
# create dict contains head 'number of fires forests in each state' and its value
record = {"Number of fires forests in each state" : dataset.groupby('state')['number'].sum()}
# visualize it as column
numberOfFiresForestByState = pd.DataFrame(record)
# draw conclusion
print("Conclusion: By data shown below, state with the most fires forest accident is Mato Grosso.")
numberOfFiresForestByState
```

```python
# split dataset
grouped = dataset.groupby('year')
# access column number from splitted dataset
numberColumn = grouped['number']
# take years from column number and put it on variable years
years = [key for key, value in numberColumn]
# take ye from column number and
totalPerYear = [value.sum() for key, value in numberColumn]
data = {'year': years,'totalPerYear': totalPerYear}
dataInFrame = pd.DataFrame(data)
print(dataInFrame.plot(x='year', ylabel= 'total', title="Data of Fires Forests in Brazil"))
print("\n")
print("Conclusion:\nFrom data shown below, in Brazil, most fires forests happened in 2002.")
```
