```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(f,df, x_start, learning_rate=0.1, n_iter=1000, tol=0.00001, decay_rate =0):
  iters= 0
  min_val = x_start
  X=[x_start]
  Y=[f(x_start)]
  diff = 0
  #for i in range(n_iter):
  while iters < n_iter:
    gradient = df(min_val)# not x but min_val
    diff= decay_rate*diff-learning_rate*gradient
    if np.all(np.abs(diff)<tol):
      break
    min_val +=diff
    iters +=1

    X.append(min_val)
    Y.append(f(min_val))
    print("Iteration",iters,"\nX value is",min_val, "\n Y value is",f(min_val))
  return min_val, X, Y
```

```python
# Question 1
#def f(x):
  #return (x+5)**2
#def df(x):
  #return 2*(x+5)
f = lambda x: (x+5)**2
df = lambda x: 2*(x+5)

x_start = 1
min_val, X, Y = gradient_descent(f, df, x_start, learning_rate=0.01, n_iter=300)
print(min_val)

```

```python
x_val=np.linspace(-10,2,400)
plt.plot(x_val, f(x_val))
for i in range(0, len(X)):
  plt.plot(X[i:i+1], Y[i:i+1], 'rx')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.show()
```
