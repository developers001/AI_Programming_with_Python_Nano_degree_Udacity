```python
#let's visualise the function
import numpy as np
import matplotlib.pyplot as plt

#Todo Question 1: Read the following gradient descent function and describe what it is doing.
def gradient_descent(function, df, x_start, learning_rate=0.1, tol=0.00001, n_iter=1000, decay_rate =0 ):
    iters = 0 #iteration counter
    min_point = x_start
    X = [x_start]
    Y = [function(x_start)]
    #Todo in question 2
    while iters < n_iter:
        diff = - learning_rate * df(min_point) #important step. #Todo in question 2
        if np.all(np.abs(diff) <= tol):
            break
        min_point += diff #important step.
        iters += 1
        X.append(min_point)
        Y.append(function(min_point))
        print("Iteration",iters,"\nX value is",min_point, "\n Y value is",function(min_point)) #Print iterations

    return min_point, X, Y

#Question 1
#Todo Question 1: initialise x_start and run gradient descent to find the minimum point
function = lambda x: (x+5)**2
df = lambda x: 2*(x+5)  #Gradient of our function
x_start = 1
ans, X, Y = gradient_descent(function, df, x_start,learning_rate=0.01, tol=0.00001, n_iter=300, decay_rate =0)
print(ans)

#Plot the curve and the lines
x = np.linspace(-10,3,500)
plt.plot(x, function(x), 'b')
for i in range(0, len(X)):
    plt.plot(X[i:i+2], Y[i:i+2], 'ro-')
plt.show()




```

```python
from decimal import Decimal, getcontext
import numpy as np

# Set the desired precision (e.g., 4 decimal places)
getcontext().prec = 4

# Rescaled functions to use Decimal for calculations
function = lambda x: Decimal(x)**4 - 5*(Decimal(x)**2) - 3*Decimal(x)
df = lambda x: 4*(Decimal(x)**3) - 10*Decimal(x) - 3

# Decay rate
def gradient_descent(function, df, x_start, learning_rate=Decimal('0.1'), tol=Decimal('0.00001'), n_iter=1000, decay_rate=Decimal('0')):
    getcontext().prec = 4
    iters = 0  # iteration counter
    min_point = Decimal(x_start)
    X = [min_point]
    Y = [Decimal(function(min_point))]
    diff = Decimal('0')

    while iters < n_iter:
        diff = Decimal(diff) * Decimal(decay_rate) - Decimal(learning_rate) * Decimal(df(min_point))  # important step.
        if abs(diff) <= tol:
            break
        min_point += diff  # important step.
        iters += 1
        X.append(min_point)
        Y.append(Decimal(function(min_point)))
        print("Iteration", iters, "\nX value is", min_point, "\n Y value is", Decimal(function(min_point)))  # Print iterations

    return min_point, X, Y

```

```python
#Generate some data
np.random.seed(42)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)

plt.clf()
plt.scatter(X, y, color='black')
plt.show()

#Implement the gradient descent function as above for linear regression for 2 variables.
def gradient_descent_reg(X, y, b_start=0.1, w_start=5, learning_rate=0.01, tol=0.00001, n_iter=1000):
    iters = 0 #iteration counter
    b = b_start
    w = w_start
    n = X.shape[0] #we have n data points
    while iters < n_iter:
        #calculate gradient of b and w
        b_gradient = np.sum(b+w*X-y)/n  # * here does element wise multiplication
        w_gradient = np.sum((b + w*X - y)*X)/n  # gradient of w
        diff_b = learning_rate * b_gradient  # difference to update b
        diff_w = learning_rate * w_gradient  # difference to update w
        if np.all(np.abs(diff_b) <= tol) and np.all(np.abs(diff_w) <= tol):
            break
        #Update the values for w and b
        b -= diff_b
        w -= diff_w

        iters += 1

    return b, w

b, w = gradient_descent_reg(X, y)

plt.scatter(X, y, color='black')
plt.plot(X, w*X + b)
plt.show()
```
